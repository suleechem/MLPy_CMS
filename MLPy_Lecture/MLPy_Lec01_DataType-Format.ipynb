> string은 ('') 혹은 ("")로 생성

> a=[10,20,30,40,50]

## Formating  
> print("나는 %s색과 %s색을 좋아해요." % ("파란","빨강"))  
> print("나는 {1}색과 {0}색을 좋아합니다.".format("파랑","빨강"))  
> print("나는 {age}살이며 {color}색을 좋아합니다.".format(age=20,color="빨강"))    
> print(f"나는 {age}살이며 {color}색을 좋아합니다.")  
> print("나는 "+ str(age) + "살이며 " + color +"색을 좋아합니다.")  


## 1. List

```python
a=[10,20,30,40,50]
print("a[0]=",a[0],", a[1]=",a[1], ", a[4]=", a[4])
print("a[-1]=",a[-1],", a[-2]=",a[-2])
```

```python
b=[10,20,'Hello',[True,3.14]]
print("b[0]=",b[0],", b[2]=",b[2],", b[3]=",b[3])
print("b[3][0]=",b[3][0])
print("b[-1][-1]=",b[-1][-1])
```

```python
c=[]
c.append(100), c.append(200), c.append(300)
print(c)
```

```python
a=[10,20,30,40,50]
print("a[0:2] ==", a[0:2])
print("a[1:] ==", a[1:])
print("a[:3] ==", a[:3])
print("a[:-2] ==", a[:-2])
print("a[:] ==", a[0:])
```

## 2. Tuple

```python
a=(10,20,30,40,50)
print("a[0]=",a[0],", a[1]=",a[1], ", a[4]=", a[4])
print("a[-1]=",a[-1],", a[-2]=",a[-2])

b=(10,20,'Hello',(True,3.14))
print("b[0]=",b[0],", b[2]=",b[2],", b[3]=",b[3])
print("b[3][0]=",b[3][0])
print("b[-1][-1]=",b[-1][-1])

print("a[0:2] ==", a[0:2])
# a[0]=100
```

```python
c=()
# c.append(100), c.append(200), c.append(300)
print(c)
```

## 3. Dictionary

```python
score={"KIM":85, "LEE":95, "JUN":90}  
print('score["KIM"] ==', score["KIM"])
print(score)
score["HAN"]=80  # 새로운 원소 추가
print(score.keys())
print(score.values())
print(score.items())
```

```python
a='A73,C'
print(a,", a[0]==", a[0])
a = a + 'D, EFG'
print(a)
b=a.split()     # separator는 빈칸 
print(b)
c=a.split(',')  # separator를 ',' 로 지정 
print(c)
```

## 4. Useful function: type( ), len( ), size( ), list( ), str( ), int( ), float( )

```python
a=[10,20,30,40,50]
b=(10,20,30,40,50) 
c={"KIM":85, "LEE":95}
d='python, shell' 
e= [[10,20], [30,40], [50,60]]
print(type(a), type(b), type(c), type(d), type(e))
print(len(a), len(b), len(c), len(d), len(e))

import numpy as np
print(np.size(e))
print(np.ndim(e))
print(np.shape(e))
```

```python
f=list(d) 
g=str(100) 
h=int(3.14) 
i=float(1)
print(f, g+"a", h, i)
print(list(c.keys()))
print(list(c.values()))
print(list(c.items()))
print(str(a))
```

## 5. Formating

```python

"""
문자열 포멧
"""

print("a"+"b")
print("a","b")

# 방법 1
print("나는 %d살 입니다." % 20)
print("나는 %s살 입니다." % 20)

print("나는 %s을 좋아합니다." % "Python")
print("Apple 은 %c로 시작합니다." % "A")

print("나는 %s색과 %s색을 좋아해요." % ("파란","빨강"))

# 방법 2
print("나는 {}살 입니다.".format(20))
print("내 생일은 {}월 {}일 입니다.".format(2,23))

print("나는 {}색과 {}색을 좋아합니다..".format("파랑","빨강"))
print("나는 {1}색과 {0}색을 좋아합니다..".format("파랑","빨강"))

# 방법 3
print("나는 {age}살이며 {color}색을 좋아합니다..".format(age=20,color="빨강"))

# 방법 4
age=43
color="파랑"
print(f"나는 {age}살이며 {color}색을 좋아합니다.")
print("나는 "+ str(age) + "살이며 " + color +"색을 좋아합니다.")
```
