## 1. List

```python
a=[10,20,30,40,50]
print("a[0]=",a[0],", a[1]=",a[1], ", a[4]=", a[4])
print("a[-1]=",a[-1],", a[-2]=",a[-2])
```

```python
b=[10,20,'Hello',[True,3.14]]
print("b[0]=",b[0],", b[2]=",b[2],", b[3]=",b[3])
print("b[3][0]=",b[3][0])
print("b[-1][-1]=",b[-1][-1])
```

```python
c=[]
c.append(100), c.append(200), c.append(300)
print(c)
```

```python
a=[10,20,30,40,50]
print("a[0:2] ==", a[0:2])
print("a[1:] ==", a[1:])
print("a[:3] ==", a[:3])
print("a[:-2] ==", a[:-2])
print("a[:] ==", a[0:])
```

## 2. Tuple

```python
a=(10,20,30,40,50)
print("a[0]=",a[0],", a[1]=",a[1], ", a[4]=", a[4])
print("a[-1]=",a[-1],", a[-2]=",a[-2])

b=(10,20,'Hello',(True,3.14))
print("b[0]=",b[0],", b[2]=",b[2],", b[3]=",b[3])
print("b[3][0]=",b[3][0])
print("b[-1][-1]=",b[-1][-1])

print("a[0:2] ==", a[0:2])
# a[0]=100
```

```python
c=()
# c.append(100), c.append(200), c.append(300)
print(c)
```

## 3. Dictionary

```python
score={"KIM":85, "LEE":95, "JUN":90}  
print('score["KIM"] ==', score["KIM"])
print(score)
score["HAN"]=80  #   
print(score.keys())
print(score.values())
print(score.items())
```

```python
a='A73,C'
print(a,", a[0]==", a[0])
a = a + 'D, EFG'
print(a)
b=a.split()     # separator  
print(b)
c=a.split(',')  # separator ','   
print(c)
```

## 4. Useful function: type( ), len( ), size( ), list( ), str( ), int( ), float( )

```python
a=[10,20,30,40,50]
b=(10,20,30,40,50) 
c={"KIM":85, "LEE":95}
d='python, shell' 
e= [[10,20], [30,40], [50,60]]
print(type(a), type(b), type(c), type(d), type(e))
print(len(a), len(b), len(c), len(d), len(e))

import numpy as np
print(np.size(e))
print(np.ndim(e))
print(np.shape(e))
```

```python
f=list(d) 
g=str(100) 
h=int(3.14) 
i=float(1)
print(f, g+"a", h, i)
print(list(c.keys()))
print(list(c.values()))
print(list(c.items()))
print(str(a))
```
