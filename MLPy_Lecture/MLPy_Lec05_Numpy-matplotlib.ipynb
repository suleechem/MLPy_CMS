## 2. Dot product
- A 행렬과 B 행렬의 행렬 곱 (dot product)는 <b>np.dot(A,B)</b>로 나타내며,  
A 행렬의 열벡터와 B 행렬의 행벡터의 크기가 같아야함.  
- 만약, 같지 않다면 reshape 또는 transpose등을 사용하여 변환해야 한다.
- $(2\times3) \cdot (3 \times 2) = (2 \times 2)$
- <b>다양한 특성을 갖는 필터 개발에 활용 </b> 
- $
(64 \times 64) \cdot (64 \times 256) \cdot (256 \times 512) \cdot (512 \times 64) \cdot (64 \times 10) = (64 \times 10)
$
<img src = "http://cmseng.skku.edu/CMSLecture/ML/img/3.png" style="max-width: 70%; height: auto;">

```python
A = np.array([[1,2,3],[4,5,6]])         # 2x3 행렬
B = np.array([[-1,-2],[-3,-4],[-5,-6]]) # 3x2 행렬 

C = np.dot(A,B)

print("A.shape ==", A.shape, ", B.shape ==",B.shape)
print("C.shape ==", C.shape)
print(C)
```

## 4.transpose
### 행과 열을 바꿈 &nbsp; <b>matrix.T </b>   
<font size=4> 
$
\begin{pmatrix}
1 & 2  \\
3 & 4 \\
5 & 6
\end{pmatrix}
\overrightarrow{transpose}
\begin{pmatrix}
1 & 2 & 3  \\
4 & 4 & 6
\end{pmatrix} 
$
</font>

```python
A = np.array([[1,2],[3,4],[5,6]])
B = A.T
print(A)
print(B)

C = np.array([1,2,3,4,5])
D = C.T
print("C.shape ==", C.shape, ", D.shape ==", D.shape)
print(D)

E = C.reshape(1,5)
F = E.T
print("E.shape ==", E.shape, ", F.shape ==", F.shape)
print(F)
```

## 6. iterator

- <b> np.nditer(A, flags=['multi_index'], op_flags=['readwrite'])</b>

```python
A = np.array([[10,20,30,40],[50,60,70,80]])
print(A, "\n")
print("A.shape ==", A.shape, "\n")

it = np.nditer(A, flags=['multi_index'], op_flags=['readwrite'])
dir(it)
# print(it.shape)
idx = it.multi_index
print(type(idx), idx)

while not it.finished:
    idx = it.multi_index
    print("index ==", idx, "value ==", A[idx])
    it.iternext()
```

## 7. Concatenate
- 행렬에 행(row) 또는 열(column)을 추가하기 위한 <b>np.concatenate</b>
- axis = 0 : 행(row) 기준 / axis = 1 : 열(column) 기준
- Machine learning 의 Regression 에서 가중치(weight)와 바이어스(bias)를 구분하지 않고   
  하나의 핼렬로 취급하기 위한 프로그램이 구현 기술

```python
A = np.array([[10,20,30],[40,50,60]])
print(A.shape)

row_add = np.array([70,80,90]).reshape(1,3)
column_add = np.array([1000,2000]).reshape(2,1)

B = np.concatenate((A, row_add), axis=0)    # 행
print(B)
B = np.concatenate((A, column_add), axis=1) # 열
print(B)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 6)
plt.rcParams['font.size'] = 10
plt.rcParams['lines.linewidth'] = 2

x = np.linspace(-1, 1, 1000)
y1 = np.sin(2 * x * np.pi)
y2 = np.cos(2 * x * np.pi)
y3 = np.ceil(3 * x)

fig, ax = plt.subplots()

ax.plot(x, y1, '--')
ax.plot(x, y2, '-.')
ax.plot(x, y3, alpha=0.7)

ax.spines['left'].set_position('center')        # 왼쪽 축을 가운데 위치로 이동
ax.spines['right'].set_visible(False)           # 오른쪽 축을 보이지 않도록
ax.spines['top'].set_visible(False)             # 위 축을 보이지 않도록
ax.spines['bottom'].set_position(('data', 0))   # 아래 축을 데이터 0의 위치로 이동
ax.tick_params('both', length=5)                # Tick의 눈금 길이 0

plt.show()
```
